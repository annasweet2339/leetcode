````python
class Solution:
    def increasingTriplet(self, nums: List[int]) -> bool:
        smallest,medium = max(nums),max(nums)
        for n in nums:
            print(smallest,medium,n)
            if n < smallest:
                smallest = n
            if smallest < n < medium:
                medium = n
            if smallest < medium < n :
                return True
        return False
````

# Time complexity O(n) Space cimplaxity:O(1)
