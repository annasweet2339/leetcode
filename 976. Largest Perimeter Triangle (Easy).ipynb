```python
class Solution:
    def largestPerimeter(self, nums: List[int]) -> int:
        nums.sort()
        for i in range(len(nums)-1,1,-1):
            if (sum(nums[i-2:i])>nums[i]): return sum(nums[i-2:i+1])  
        return 0
```

# Time Complexity: O(nlogn)     Space complexity: O(1)

.sort() in python is done in O nlogn time and O1 space

Final complexity would be O(nlong +n) for time and O(1) for space, but since O(nlogn) takes over O(n), you could just say your time complexity is O(nlogn) and space O(1).
