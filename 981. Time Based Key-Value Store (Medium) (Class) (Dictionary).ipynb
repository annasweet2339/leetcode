````python
class TimeMap:

    def __init__(self):
        self.timestamp_list = {}

    def set(self, key: str, value: str, timestamp: int) -> None:
        self.timestamp_list[timestamp] = {key:value}
        print('set timestamp_list', self.timestamp_list)
        
    def get(self, key: str, timestamp: int) -> str:
        for t in range(timestamp,0,-1):
            if t in self.timestamp_list.keys():
                if key in self.timestamp_list[t]:
                    print('get',key,timestamp,':',self.timestamp_list[t][key])
                    return self.timestamp_list[t][key]
        print('get',key,timestamp,': ""')
        return ""
                
# Your TimeMap object will be instantiated and called as such:
# obj = TimeMap()
# obj.set(key,value,timestamp)
# param_2 = obj.get(key,timestamp)
````
